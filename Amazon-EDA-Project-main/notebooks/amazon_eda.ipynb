## Data Cleaning

**null values**

```python
amazon.isnull().sum()
```

**handel null values**

```python
amazon.dropna(inplace=True)
```

```python
amazon.shape
```

**Fix data types**

```python
amazon['discounted_price']=amazon['discounted_price'].str.replace('₹','')
amazon['discounted_price']=amazon['discounted_price'].astype(str).str.replace(',','').astype(float)
```

```python
amazon['actual_price']=amazon['actual_price'].str.replace('₹','')
amazon['actual_price']=amazon['actual_price'].str.replace('-','')
amazon['actual_price']=amazon['actual_price'].astype(str).str.strip()
amazon['actual_price']=amazon['actual_price'].astype(str).str.replace(',','').astype(float)
```

```python
amazon['discount_percentage']=amazon['discount_percentage'].astype(str).str.replace('%','').astype(float)
```

```python
amazon['rating']=amazon['rating'].astype(str).str.strip()
amazon['rating']=amazon['rating'].str.replace('|','')
amazon['rating']=amazon['rating'].str.replace('/','')
amazon['rating']=amazon['rating'].str.replace('**','')
amazon['rating']=amazon['rating'].str.replace(' ','')
amazon['rating'] = pd.to_numeric(amazon['rating'])
```

```python
amazon['rating_count']=amazon['rating_count'].astype(str).str.replace(',','').astype(float)
```

## Data Visualization

**Category distribution charts**

```python
category_distributed=amazon['category'].value_counts().head(10)
plt.Figure(figsize=(10,8))
plt.bar(category_distributed.index,category_distributed.values , color='r')
plt.xlabel("category")
plt.ylabel("product")
plt.title("Number of Product")
plt.xticks(rotation=45,ha='right')
plt.show()
```

Computers&Accessories|Accessories&Peripherals|Cables : this category have max number of products

**Price vs Discount analysis**

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=amazon,x='actual_price',y='discounted_price' ,alpha=0.6, color='orange')
plt.xlabel("actual price")
plt.ylabel("discount price")
plt.title("price vs discount")
plt.show()
```

if price increase , discount will increase

**Rating distribution**

```python
plt.figure(figsize=(10,8))
sns.barplot(data=amazon.head(),x='product_name',y='rating')
plt.title("Rating")
plt.xlabel("product name")
plt.ylabel("rating")
plt.xticks(rotation=90,ha='right')
plt.show()
```

""wayona , boAt , portronics"" have max rate

**Top-rated products**
